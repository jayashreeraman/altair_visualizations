```python
import altair as alt
from vega_datasets import data

```

```python
disasters = data.disasters.url

print(disasters)
```

```python
#enable the altair rendered
alt.renderers.enable('notebook')

# draw the chart
disasters = data.disasters.url
alt.Chart(disasters).mark_point().encode(
    x='Year:T',
    y='Deaths:Q',
    color='Entity:N'
)
```

```python
# draw the cars plot : with altair 
cars = data.cars.url
alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q'
)
```

```python
# define the theme by returning the dictionary of configurations
def black_marks():
    return {
        'config': {
            'view': {
                'height': 300,
                'width': 400,
            },
            'mark': {
                'color': 'black',
                'fill': 'green'
            }
        }
    }
```

```python
# register the custom theme under a chosen name
alt.themes.register('black_marks', black_marks)
# enable the newly registered theme
alt.themes.enable('black_marks')
# draw the chart
cars = data.cars.url
alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q'
)
```

```python
#Define a custom theme
def custom_theme():
    return {
        'config': {
            'view': {
                'height': 300,
                'width': 400,
            },
            'mark': {
                'color': 'red',
                'fill': '#fbdecd',
            },
            'axisBottom': {
                'labelFontSize': 30,
            },
        }
    }
# register the custom theme under a chosen name
alt.themes.register('custom_theme', custom_theme)
# enable the newly registered theme
alt.themes.enable('custom_theme')
```

```python

#Plot with custome theme
cars = data.cars.url
alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q'
)
```
